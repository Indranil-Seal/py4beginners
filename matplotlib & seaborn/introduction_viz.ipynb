# In this script:

Matplotlib is used to create a line plot and a scatter plot.
Seaborn is used to create a distribution plot and a pair plot.
>The plt.show() function is used to display each plot. Adjust the figure sizes, colors, and other parameters based on your preferences. If you haven't installed Seaborn yet, you can do so using pip install seaborn. Feel free to run this script in a Python environment to see the plots.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Matplotlib: Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)')
plt.title('Matplotlib Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Matplotlib: Scatter Plot
np.random.seed(42)
x_scatter = np.random.rand(50)
y_scatter = 2 * x_scatter + 1 + 0.1 * np.random.randn(50)
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(x_scatter, y_scatter, label='Data Points')
plt.title('Matplotlib Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Seaborn: Distribution Plot
data_distribution = np.random.randn(1000)
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data_distribution, kde=True, color='skyblue')
plt.title('Seaborn Distribution Plot')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Seaborn: Pair Plot
iris = sns.load_dataset('iris')

plt.figure(figsize=(8, 6))
sns.pairplot(iris, hue='species')
plt.title('Seaborn Pair Plot')
plt.show()
```
